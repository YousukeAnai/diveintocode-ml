```python
#問題　円の体積を算出する関数
#
#太陽系から最も遠い星「エリス」の軌道長半径を太陽系の半径とし、太陽系を円として体積を計算　
#https://ja.wikipedia.org/wiki/太陽系#太陽系の境界      ＿エリスの軌道長半径確認(67.664au)
#https://allabout.co.jp/gm/gc/469416/                   ＿1AU≒1億5千万km
#
import matplotlib.pyplot as plt
import numpy as np
import math

def circle_volume(def_r):
    """
    def circle_volume Summary
        円の体積を算出する関数。
        半径(m)を入力すると、円の体積(m3)、を返す。

    Parameteres
        Input
            def_r (float) : 円の半径
        Return
            def_circle_volume (float) : 円の体積
    """
    PAI = math.pi
    def_circle_volume = (4 * PAI * def_r**3) / 3
    return def_circle_volume
```

```python
#問題　バイバインの理屈で、饅頭の体積が目標体積を超えるまでの時間、を算出する関数
#栗饅頭を半径2.5cmの円形とする。
#
MANJUU_VOLUME_0_R = 0.025
TIME_TO_DUBLE = 5
def baibain_manjuu_volume(def_target_volume_list):
    def_time_list = [0]
    def_manjuu_volume_list = [circle_volume(MANJUU_VOLUME_0_R)]
    def_result_time_list = []
    def_result_manjuu_volume_list = []
    print("{}分後、栗饅頭の体積：{:.4g}m3".format(def_time_list[-1],def_manjuu_volume_list[-1]))
    for i in def_target_volume_list:
        while def_manjuu_volume_list[-1] < i:
            def_time_list.append(def_time_list[-1] + TIME_TO_DUBLE)
            def_manjuu_volume_list.append(def_manjuu_volume_list[-1]*2)
            print("{}分後、栗饅頭の体積：{:.4g}m3".format(def_time_list[-1],def_manjuu_volume_list[-1]))
        def_result_time_list.append(def_time_list[-1])
        def_result_manjuu_volume_list.append(def_manjuu_volume_list[-1])
    return def_result_time_list, def_result_manjuu_volume_list, def_time_list, def_manjuu_volume_list
```

```python
#　問題　＿各関数を用いて、バイバインの饅頭体積が東京ドーム、太陽系の体積になるまでの時間を算出し、グラフで表す。
AU_PER_METER = 1.5E+8  #
ERIS_OF_ORBIT_MAJOR_AXIS_PER_AU = 67.664  #
SOLER_SYSTEM_R = ERIS_OF_ORBIT_MAJOR_AXIS_PER_AU * AU_PER_METER  #

TOKYO_DOME_VOLUME = 1240000  #https://ja.wikipedia.org/wiki/東京ドーム_(単位)

target_volume_list = [TOKYO_DOME_VOLUME, circle_volume(SOLER_SYSTEM_R)]
target_name_list_j = ["東京ドーム","太陽系"]
target_name_list_e = ["TOKYO_DOME","SOLAR_SYSTEM"]

result_time_list, result_manjuu_volume_list, time_list, manjuu_volume_list = baibain_manjuu_volume(target_volume_list)

list2graf(time_list, manjuu_volume_list, "Volume_of_Manjuu", 
         [0,result_time_list[-1]+10],[target_volume_list[0], target_volume_list[0]], target_name_list_e[0],
         [0,result_time_list[-1]+10],[target_volume_list[1], target_volume_list[1]], target_name_list_e[1],
         "TIME_vs_MANJUU_VOLUME", "TIME[m]", "VOLUME[m3]")

for target_name_j, target_volume, result_time, result_manjuu_volume in \
zip(target_name_list_j, target_volume_list, result_time_list, result_manjuu_volume_list):
    print("------------------------\n饅頭が{}({:.3g}m3)を覆うまでの時間は{}分後で、その時の饅頭の体積は{:.3g}m3"
    .format(target_name_j, target_volume, result_time, result_manjuu_volume))
```
