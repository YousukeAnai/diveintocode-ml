```python
#太陽系から最も遠い星「エリス」の軌道長半径を太陽系の半径とし、太陽系を円として体積を計算　
#https://ja.wikipedia.org/wiki/太陽系#太陽系の境界      ＿エリスの軌道長半径確認(67.664au)
#https://allabout.co.jp/gm/gc/469416/                   ＿1AU≒1億5千万km
#栗饅頭を半径を2.5cmの円とする。

# T      0       5       10         15
# M_V    M_V0   M_V0*2   M_V0*2*2
import math
math.pi
```

```python
import matplotlib.pyplot as plt
import numpy as np

Solar_System_Volume = (4 * math.pi * (67.664*1.5E+8)**3)/3
Manjuu_Volume_0 = (4 * math.pi * (0.025)**3)/3

def target_volume(v):
    t_list = [0]
    manjuu_volume_list = [Manjuu_Volume_0]
    while manjuu_volume_list[-1] < v:
#        print("{}分後、栗饅頭の体積：{:.4g}m3".format(t_list[-1],manjuu_volume_list[-1]))
        t_list.append(t_list[-1]+5)
        manjuu_volume_list.append(manjuu_volume_list[-1]*2)
    return t_list,manjuu_volume_list
```

```python
a, b =target_volume(Solar_System_Volume)

print("饅頭が太陽系(体積{:.4g}m3)を覆うまでの時間は{}分後".format(Solar_System_Volume,a[-1]))
print("そのときの饅頭の体積：{:.4g}m3".format(b[-1]))
```

```python
t_list,manjuu_volume_list = target_volume(Solar_System_Volume)
%matplotlib inline
X = t_list
Y = manjuu_volume_list
plt.title("TIME vs MANJUU_VOLUME")
plt.xlabel("TIME [m]")
plt.ylabel("MANJUU_VOLUME")
plt.plot(X,Y)
plt.show()
```
