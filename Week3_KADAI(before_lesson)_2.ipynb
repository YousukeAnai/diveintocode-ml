### 【問題6】分布の確認 
seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
目的変数に対して対数変換を行う。
対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。
対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

```python
#seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
import scipy
sns.distplot(df["LotArea"], kde=False, rug=False, bins=20) 
print("歪度：{}".format(scipy.stats.skew(df["LotArea"])))
print("尖度：{}".format(scipy.stats.kurtosis(df["LotArea"])))
```

```python
#目的変数に対して対数変換を行う。
import math
df_log_lotarea = df["LotArea"].map(math.log10)
```

```python
#対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。
sns.distplot(df_log_lotarea, kde=False, rug=False, bins=20) 
print("歪度：{}".format(scipy.stats.skew(df_log_lotarea)))
print("尖度：{}".format(scipy.stats.kurtosis(df_log_lotarea)))
```

```python
#対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。
```

```python
変換前は、歪度が非常に大きく、分布に偏りがある。また尖度が非常に大きい。
変換後は、対数化することで歪度が小さくなり左右対称に近づき、尖度も変換前より小さくなり適度にばらついて正規分布に近づいている。
```
