```python
#　問題２　＿結果算出

def thickness_function(def_target, def_thickness=0.00008):
    """
    def thickness_function Summary
        紙を折った時の紙の厚さが、目標値に達するためには何回折ればよいかを算出する関数。
        
        目標値(単位:m)を入力し、
        折った回数毎に、紙の厚さ、を出力し、
        下記Returnsに記載された項目、を返す。

    Parameteres
        Input
            def_target (float) : 目標値
            def_thickness (float) : 折る前の紙の厚さ(default=0.00008)

        Return
            n (int) : 紙を折った回数
            def_folded_thickness_list (list) : 紙の厚さの折った回数毎のリスト
        
        other
            def_folded_thickness (float) : 折った後の紙の厚さ
        
    Appendix
        while文にて紙の厚さが目標値より大きくなるまで、紙の厚さを2倍にすることを繰り返す。
        紙を折るごとに、def_folded_thickness_listに紙の厚さを追加していく。
    """
    def_folded_thickness = def_thickness
    def_folded_thickness_list = [def_thickness]
    
    n = 0
    while def_folded_thickness < def_target:
        def_folded_thickness *= 2
        def_folded_thickness_list.append(def_folded_thickness)
        n += 1
        print("折った回数：{}回、厚さ：{:.3g}m".format(n, def_folded_thickness))
    return n,def_folded_thickness_list

#https://ja.wikipedia.org/wiki/近い恒星の一覧    ＿最も近い恒星調査 (プロキシマ)
#https://allabout.co.jp/gm/gc/469416/　　　　　　＿1光年を調査
DISTANCE_OF_EARTH_TO_PUROKISIMA_PER_KOUNEN = 4.243 
METER_PER_KOUNEN = 9.5E+15
DISTANCE_OF_EARTH_TO_PUROKISIMA_PER_METER = DISTANCE_OF_EARTH_TO_PUROKISIMA_PER_KOUNEN * METER_PER_KOUNEN

num_of_folded, folded_thickness_list = thickness_function(DISTANCE_OF_EARTH_TO_PUROKISIMA_PER_METER)

print("----------------------\n厚さが{:.3g}m(最も近い恒星(プロキシマ)への距離)に達するには紙を{}回折る必要がある"
      .format(DISTANCE_OF_EARTH_TO_PUROKISIMA_PER_METER, num_of_folded))
```

```python
import math
def thickness_function(def_target, def_thickness=0.00008):
    """
    def thickness_function Summary
        紙を折った時の紙の厚さが、目標値に達するためには何回折ればよいかを算出する関数。
        
        目標値(単位:m)を入力し、
        折った回数毎に、紙の厚さ、を出力し、
        下記Returnsに記載された項目、を返す。

    Parameteres
        Input
            def_target (float) : 目標値
            def_thickness (float) : 折る前の紙の厚さ(default=0.00008)

        Return
            n (int) : 紙を折った回数
            def_folded_thickness_list (list) : 紙の厚さの折った回数毎のリスト
        
        other
            def_folded_thickness (float) : 折った後の紙の厚さ
        
    Appendix
        while文にて紙の厚さが目標値より大きくなるまで、紙の厚さを2倍にすることを繰り返す。
        紙を折るごとに、def_folded_thickness_listに紙の厚さを追加していく。
    """
    def_folded_thickness = def_thickness
    def_folded_thickness_list = [def_thickness]
    
    n = 0
    while def_folded_thickness < def_target:
        def_folded_thickness *= 2
        def_folded_thickness_list.append(def_folded_thickness)
        n += 1
        
    pai = math.pi
    def_required_paper_length = ((pai*def_folded_thickness_list[0]) * (2**n + 4) * (2**n - 1)) / 6
    
    return n, def_folded_thickness_list, def_required_paper_length

MT_FUJI = 3776
DISTANCE_OF_MOON_TO_EARTH_PER_METER = 384400000
target_value_list = [MT_FUJI, DISTANCE_OF_MOON_TO_EARTH_PER_METER, DISTANCE_OF_EARTH_TO_PUROKISIMA_PER_METER]
target_name_list = ["富士山の高さ","月までの距離","プロキシマ(太陽の次に近い恒星)までの距離"]

for name, value in zip(target_name_list, target_value_list):
    num_of_folded, folded_thickness_list, required_paper_length = thickness_function(value)
    print("----------------------\n厚さが{:.4g}m({})に達するには、\n紙を{}回折る必要があり、\nそれに必要な紙の長さは{:.3g}m"
      .format(value, name, num_of_folded, required_paper_length))
               
```
