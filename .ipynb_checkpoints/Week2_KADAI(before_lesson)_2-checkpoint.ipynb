```python
#問題１
import numpy as np
import matplotlib.pyplot as plt

def func_2dim_covariance_matrix(def_mean, def_covariance_matrix, def_num_plot):
    def_out_covariance_matrix = np.random.multivariate_normal(def_mean, def_covariance_matrix , def_num_plot)
    return def_out_covariance_matrix

average = (-3, 0)
covariance_matrix = [[1.0,0.8],[0.8,1.0]]
num_plot = 500
np.random.seed(0)
out_covariance_matrix = func_2dim_covariance_matrix(average, covariance_matrix, num_plot)
print(out_covariance_matrix)
print(out_covariance_matrix.shape)
```

```python
#問題２
def func_scatter_2dim_covariance_matrix(def_2dim_covariance_matrix):
    x=def_2dim_covariance_matrix[:,0]
    y=def_2dim_covariance_matrix[:,1]
    plt.scatter(x,y)
    plt.show()
np.random.seed(0)
func_scatter_2dim_covariance_matrix(out_covariance_matrix)
```

```python
#問題３
def func_histgram_2dim_covariance_matrix(def_input_hist_data, def_bins, def_xlim,
                                         def_title, def_xlabel, def_ylabel):
    plt.hist(def_input_hist_data, def_bins, alpha=1)
    plt.title(def_title)
    plt.xlabel(def_xlabel)
    plt.ylabel(def_ylabel)
    plt.xlim(def_xlim)
    plt.show()

np.random.seed(0)
BINS = 500
X_LIMIT = (-5,0)
TITLE = "histogram of x1"
X_LABEL = "x1"
Y_LABEL = "frequency"
input_hist_data1 = out_covariance_matrix[:,0]
histgram_dim1 = func_histgram_2dim_covariance_matrix(input_hist_data1, BINS, X_LIMIT, TITLE, X_LABEL, Y_LABEL)

TITLE = "histogram of x2"
X_LABEL = "x2"
input_hist_data2 = out_covariance_matrix[:,1]
histgram_dim2 = func_histgram_2dim_covariance_matrix(input_hist_data2, BINS, X_LIMIT, TITLE, X_LABEL, Y_LABEL)
```

```python
#問題４
average2 = (0, -3)
covariance_matrix = [[1.0,0.8],[0.8,1.0]]
num_plot = 500
np.random.seed(1)
out_covariance_matrix2 = func_2dim_covariance_matrix(average2, covariance_matrix, num_plot)

def func_2scatter_2dim_covariance_matrix(def_2dim_covariance_matrix, def_2dim_covariance_matrix2,
                                         def_title, def_xlabel, def_ylabel):
    x=def_2dim_covariance_matrix[:,0]
    y=def_2dim_covariance_matrix[:,1]
    x2=def_2dim_covariance_matrix2[:,0]
    y2=def_2dim_covariance_matrix2[:,1]
    plt.title(def_title)
    plt.xlabel(def_xlabel)
    plt.ylabel(def_ylabel)
    plt.scatter(x,y,color="b",label="0")
    plt.scatter(x2,y2,color="g",label="1")
    plt.legend()
    plt.show()

TITLE = "2scatter"
X_LABEL = "x1"
Y_LABEL = "x2"
func_2scatter_2dim_covariance_matrix(out_covariance_matrix, out_covariance_matrix2, TITLE, X_LABEL, Y_LABEL)
```

```python
#問題５
out_covariance_matrix_ketugou = np.concatenate((out_covariance_matrix, out_covariance_matrix2), axis=0)
print(out_covariance_matrix_ketugou.shape)
func_scatter_2dim_covariance_matrix(out_covariance_matrix_ketugou)
```
