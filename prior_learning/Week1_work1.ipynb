### 【問題1】100日後の米粒の数
　100日後には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
#　問題１　＿計算
import matplotlib.pyplot as plt
import numpy as np

day_list = [0]
get_rice_list = [0.5]
total_rice_list = [0]

for day in range(1,101):
    day_list.append(day)
    get_rice_list.append(get_rice_list[-1]*2)
    total_rice_list.append(get_rice_list[-1]+total_rice_list[-1])
    print("{}日目、もらう米：{:.3g}粒、もらった米合計：{:.3g}粒".format(day_list[-1],get_rice_list[-1],total_rice_list[-1]))
```

```python
#　問題１　＿グラフ
%matplotlib inline
X = day_list
Y = total_rice_list
Y2 = get_rice_list
plt.title("DAY vs TOTAL_RICE,GET_RICE")
plt.xlabel("DAY")
plt.ylabel("TOTAL_RICE , GET_RICE")
plt.grid()
plt.plot(X,Y,color="r",marker='o',label="TOTAL_RICE")
plt.plot(X,Y2,color="b",marker='x',label="GET_RICE")
plt.legend()
plt.show()

X_g2 = day_list[90:]
Y_g2 = total_rice_list[90:]
Y2_g2 = get_rice_list[90:]
plt.title("ZOOM____DAY vs TOTAL_RICE,GET_RICE")
plt.xlabel("DAY")
plt.ylabel("TOTAL_RICE , GET_RICE")
plt.grid()
plt.plot(X_g2,Y_g2,color="r",marker='o',label="TOTAL_RICE")
plt.plot(X_g2,Y2_g2,color="b",marker='x',label="GET_RICE")
plt.legend()
plt.show()
print("-----------------\n約94から急上昇している")
```

### 【問題2】100日後以外の米粒の数
この逸話は30日だったり81日だったりと日数が曖昧です。どの話にも対応できるように日数を引数とした 関数 を作成しましょう。関数を作成した上で、コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。前の問題同様に折れ線グラフも出力してください。

以下の雛形には関数としての枠組みだけではなく、引数や返り値についてのコメントも記載してあります。参考にして、自分なりにも分かりやすくコメントを書き換えてください。

```python
#　問題２　＿関数
def day_function(def_target_day):
    """
    def day_function Summary
        曽呂利新左衛門の米の逸話を検証する関数。
        
        米をもらう期間(単位:日)を入力し、
        日ごとに、もらった米粒、その日までに貰った米粒の合計、を出力し、
        下記Returnsに記載された項目、を返す。

    Parameteres
        def_target_day (int) : 米をもらう期間
        def_day (int) : 米を貰う日(for文で各日を表す変数として使用)

    Returns
        def_day_list (list) : 米を貰う日数のリスト
        def_get_rice_list (list) : 貰った米粒の日毎のリスト
        def_total_rice_list (list) : 貰った米粒の日毎のリスト

    Appendix
        for文にて1日目から最終日までの各米の計算を実施。
        def_get_rice_list、def_total_rice_list　にて1つ前のリストの値を用いているおり、1日目の計算でエラーが出ないように
        下記にてdef_day_list、def_get_rice_list、def_total_rice_list初期値を設定。
        def_get_rice_listの初期値は1日目が1になるように、0.5としている。
    """
    def_day_list = [0]
    def_get_rice_list = [0.5]
    def_total_rice_list = [0]

    for def_day in range(1,def_target_day+1):
        def_day_list.append(def_day)
        def_get_rice_list.append(def_get_rice_list[-1]*2)
        def_total_rice_list.append(def_get_rice_list[-1]+def_total_rice_list[-1])
        print("{}日目、もらう米：{:.3g}粒、もらった米合計：{:.3g}粒".format(def_day_list[-1],def_get_rice_list[-1],def_total_rice_list[-1]))
    return def_day_list,def_get_rice_list,def_total_rice_list

day_list,get_rice_list,total_rice_list = day_function(120)

print("-----------------\n結論＿本コースを受講する日数{}日間でもらう米粒の合計：{:.3g}粒".format(day_list[-1],total_rice_list[-1]))
```

```python
#　問題２　＿グラフ
%matplotlib inline
X = day_list
Y = total_rice_list
Y2 = get_rice_list
plt.title("DAY vs TOTAL_RICE,GET_RICE")
plt.xlabel("DAY")
plt.ylabel("TOTAL_RICE , GET_RICE")
plt.grid()
plt.plot(X,Y,color="r",marker='o',label="TOTAL_RICE")
plt.plot(X,Y2,color="b",marker='x',label="GET_RICE")
plt.legend()
plt.show()

X_g2 = day_list[len(day_list)-10:]
Y_g2 = total_rice_list[len(day_list)-10:]
Y2_g2 = get_rice_list[len(day_list)-10:]
plt.title("ZOOM____DAY vs TOTAL_RICE,GET_RICE")
plt.xlabel("DAY")
plt.ylabel("TOTAL_RICE , GET_RICE")
plt.grid()
plt.plot(X_g2,Y_g2,color="r",marker='o',label="TOTAL_RICE")
plt.plot(X_g2,Y2_g2,color="b",marker='x',label="GET_RICE")
plt.legend()
plt.show()
print("-----------------\n約144から急上昇している")
```

### 【問題3】何人が何日生活できる米か¶
この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。

関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。

ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
#　問題３　＿関数
import math

def day_function(def_target_day,def_num_man):
    """
    def day_function Summary
        曽呂利新左衛門の逸話の理屈で、何人が何日生活できるかを検証する関数。
        
        米をもらう期間(単位:日)、人数、を入力し、
        日ごとに、もらった米粒、その日までに貰った米粒の合計、貰った米に対し、その人数で生活できる日数を出力し、
        下記Returnsに記載された項目、を返す

    Parameteres
        def_target_day (int) : 米をもらう期間(default:100)
        def_day (int) : 米を貰う日(for文で各日を表す変数として使用)
        def_men_live_days (int) :  入力した人数が貰った米の合計で生きれる日数
        
        LIVE_1DAY_PER_KCAL (float) : 1日生きるのに必要なカロリー(kcal)
                                        参照元：https://www.med.or.jp/forest/health/eat/01.php   :30-49歳の男が普通の生活をする場合で計算
        KCAL_PER_GRAM (float) : 1キロカロリーが米何gに相当するか
                                    参照元：https://calorie.slism.jp/101088/  
        GRAM_PER_TUBU (float) : 米1粒が何gか
                                    参照元：https://ameblo.jp/fenegi/entry-10598490252.html
        
        LIVE_1DAY_PER_RICE_GRAINS (float) : 1人が1日生きるのに必要な米粒

    Returns
        def_day_list (list) : 米を貰う日数のリスト
        def_get_rice_list (list) : 貰った米粒の日毎のリスト
        def_total_rice_list (list) : 貰った米粒の日毎のリスト
        def_men_live_days_list (list) : 入力した人数が貰った米の合計で生きれる日数の日毎のリスト
        def_num_man (int) : 入力した人数
        
    Appendix
        for文にて1日目から最終日までの各米の計算を実施。
        def_get_rice_list、def_total_rice_list　にて1つ前のリストの値を用いているおり、1日目の計算でエラーが出ないように
        下記にてdef_day_list、def_get_rice_list、def_total_rice_list初期値を設定。
        def_get_rice_listの初期値は1日目が1になるように、0.5としている。
    """
    LIVE_1DAY_PER_KCAL = 2677.5
    KCAL_PER_GRAM = 1.68
    GRAM_PER_TUBU = 0.025
    LIVE_1DAY_PER_RICE_GRAINS = LIVE_1DAY_PER_KCAL/KCAL_PER_GRAM/GRAM_PER_TUBU
    def_all_live_1day_per_rice_grains = LIVE_1DAY_PER_RICE_GRAINS*def_num_man
    
    def_day_list = [0]
    def_get_rice_list = [0.5]
    def_total_rice_list = [0]
    def_men_live_days_list =[0]
    
    
    for def_day in range(1,def_target_day+1):
        def_day_list.append(def_day)
        def_get_rice_list.append(def_get_rice_list[-1]*2)
        def_total_rice_list.append(def_get_rice_list[-1]+def_total_rice_list[-1])
        def_men_live_days = math.floor(def_total_rice_list[-1]/def_all_live_1day_per_rice_grains)
        def_men_live_days_list.append(def_men_live_days)
        print("{}日目、もらう米：{:.3g}粒、もらった米合計：{:.3g}粒、{}人が生きれる日数：{:.3g}日"
              .format(def_day_list[-1], def_get_rice_list[-1], def_total_rice_list[-1], def_num_man, def_men_live_days_list[-1]))
    return def_day_list, def_get_rice_list, def_total_rice_list, def_men_live_days_list, def_num_man

day_list, get_rice_list, total_rice_list, men_live_days_list, num_man = day_function(120,12)

print("----------------------------------------")
print("結論＿本コースを受講する日数{}日間でもらう米粒の合計：{:.3g}粒".format(day_list[-1],total_rice_list[-1]))
print("      その米で、{}人が生活できる日数は{:.3g}日".format(num_man,men_live_days_list[-1]))
```

```python
#　問題３　＿グラフ
%matplotlib inline
X = day_list
Y = total_rice_list
Y2 = get_rice_list
plt.title("DAY vs TOTAL_RICE")
plt.xlabel("DAY")
plt.ylabel("TOTAL_RICE , GET_RICE")
plt.grid()
plt.plot(X,Y,color="r",marker='o',label="TOTAL_RICE")
plt.plot(X,Y2,color="b",marker='x',label="GET_RICE")
plt.legend()
plt.show()

X_g2 = day_list[len(day_list)-10:]
Y_g2 = total_rice_list[len(day_list)-10:]
Y2_g2 = get_rice_list[len(day_list)-10:]
plt.title("ZOOM____DAY vs TOTAL_RICE,GET_RICE")
plt.xlabel("DAY")
plt.ylabel("TOTAL_RICE , GET_RICE")
plt.grid()
plt.plot(X_g2,Y_g2,color="r",marker='o',label="TOTAL_RICE")
plt.plot(X_g2,Y2_g2,color="b",marker='x',label="GET_RICE")
plt.legend()
plt.show()

Y3 = men_live_days_list
plt.title("DAY vs MEN_LIVE_DAYS")
plt.xlabel("DAY")
plt.ylabel("MEN_LIVE_DAYS [DAY]")
plt.grid()
plt.plot(X,Y3,color="g",marker='+',label="MEN_LIVE_DAY")
plt.legend()
plt.show()

X_g4 = day_list[len(day_list)-10:]

Y3_g4 = men_live_days_list[-10:]
plt.title("ZOOM____DAY vs MEN_LIVE_DAYS")
plt.xlabel("DAY")
plt.ylabel("MEN_LIVE_DAYS [DAY]")
plt.grid()
plt.plot(X_g4,Y3_g4,color="g",marker='+',label="MEN_LIVE_DAY")
plt.legend()
plt.show()


print("-----------------\n約144から急上昇している")
```
