### 【問題2】学習と検証 
データを簡単に分析、前処理し、学習、検証するまでの一連の流れを作成・実行してください。
検証にはこのコンペティションで使用される評価指標を用いるようにしてください。学習に用いる手法は指定しません。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from sklearn.datasets import load_iris
df = pd.read_csv(r"C:\Users\anai\dive\Dataset\application_train.csv")
df.head()
```

```python
#欠損をすべて消す。
df.isnull().sum()
df_drop = df.dropna()
df_drop.isnull().sum()
df_drop.shape
```

```python
X_ = df_drop.drop("TARGET",axis=1)
y_ = df_drop.loc[:,["TARGET"]]
```

```python
X_.shape
y_.shape
```

```python
X_2 = X_.select_dtypes(exclude="object")
X_2.info()
```

```python
from sklearn.model_selection import train_test_split
X = np.array(X_2)
y = np.array(y_)
display(X_.shape)
display(y_.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=0)
display(X_train.shape)
display(X_test.shape)
display(y_train.shape)
display(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_train, y_train) 

print("accuracy_score：{:.3g}".format(accuracy_score(y_test, neigh.predict(X_test))))
print("precision_score：{:.3g}".format(precision_score(y_test, neigh.predict(X_test))))
print("recall_score：{:.3g}".format(recall_score(y_test, neigh.predict(X_test))))
print("f1_score：{:.3g}".format(f1_score(y_test, neigh.predict(X_test))))
from sklearn.metrics import confusion_matrix
print("confusion_matrix：\n{}".format(confusion_matrix(y_test, neigh.predict(X_test))))
```

```python
y_pred = neigh.predict(X_test)
```

```python
from sklearn import metrics
```

```python
fpr, tpr, thresh = metrics.roc_curve(y_test, y_pred)
auc = metrics.auc(fpr, tpr)
auc
```

```python
#TraindﾃﾞｰﾀとTestﾃﾞｰﾀをくっつける。　TARGETは外しておく。
df_train_no_y = df_train.drop("TARGET",axis=1)
df = pd.concat([df_train_no_y, df_test])
df.shape
```

```python
#Target以外の特徴量の相関関係を確認。まずはint型から。
#要素数が多い"CNT_CHILDREN", "DAYS_EMPLOYED", "DAYS_ID_PUBLISH", "HOUR_APPR_PROCESS_START"の相関関係を確認
import seaborn as sns
df_int = df.select_dtypes(include="int64")
df_int_select = df_int.loc[:,["CNT_CHILDREN", "DAYS_EMPLOYED", "DAYS_ID_PUBLISH", "HOUR_APPR_PROCESS_START"]]
df_int_select_corr = df_int_select.corr()
sns.heatmap(df_int_select_corr, annot=True, fmt="1.1f")
```

```python
#Targetをくっつけて最終Trainﾃﾞｰﾀを作る。
df_train3 = df_train.loc[:,["SK_ID_CURR","TARGET"]].merge(df_1hot, on=["SK_ID_CURR"])
display(df_train3.shape)
```

```python
from sklearn.model_selection import train_test_split
X_df_train3 = df_train3.drop("TARGET",axis=1)
y_df_train3 = df_train3.loc[:,["TARGET"]]

X = X_df_train3.values
y = y_df_train3.values

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.8, 
                                                    random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
def func_models_eval2(_X_train, _X_test, _y_train, _y_test):

    models = {
         "KNeighborsClassifier": KNeighborsClassifier(n_neighbors=5),
     
         "LogisticRegression":LogisticRegression(random_state=0),
        
         "SVC":SVC(gamma="auto"),

         "DecisionTree":DecisionTreeClassifier(random_state=0),

         "RandomForest":RandomForestClassifier(random_state=0)
    }
    
    _df_results = pd.DataFrame()
    for model_name, model in models.items():
        model.fit(_X_train, _y_train)
        fpr, tpr, thresh = metrics.roc_curve(_y_test, model.predict(_X_test))
        _df_result = pd.DataFrame({model_name:[metrics.auc(fpr, tpr)]}, 
                                index=["auc_score"])
        _df_results = pd.concat([_df_results, _df_result],axis=1)                        
    return _df_results
    
df_results = func_models_eval2(X_train, X_test, y_train, y_test)
df_results
```
