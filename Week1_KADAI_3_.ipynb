```python
#問題　円の体積を算出する関数
#
#太陽系から最も遠い星「エリス」の軌道長半径を太陽系の半径とし、太陽系を円として体積を計算　
#https://ja.wikipedia.org/wiki/太陽系#太陽系の境界      ＿エリスの軌道長半径確認(67.664au)
#https://allabout.co.jp/gm/gc/469416/                   ＿1AU≒1億5千万km


import matplotlib.pyplot as plt
import numpy as np
import math

def circle_volume(def_r):
    """
    def circle_volume Summary
        円の体積を算出する関数。
        半径(m)を入力すると、円の体積(m3)、を返す。

    Parameteres
        Input
            def_r (float) : 円の半径
        Return
            def_circle_volume (float) : 円の体積
    """
    PAI = math.pi
    def_circle_volume = (4 * PAI * def_r**3) / 3
    return def_circle_volume
```
